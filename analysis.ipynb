## Step 1: Introduction

In this dataset, we explore recorded major power outages in the continental U.S. spanning from January 2000 to July 2026. Taken from Purdue University, the *Major Power Outage Risks in the U.S.*  dataset includes information on the location of the outages, regional climate features, land-use, electricity consumption, and economic characteristics of the states. Given these details, what might cause an area to be of high risk, where risk is defined by it's duration? Does seasonality have an impact on the severity of the types of outages that ensue? More broadly, what are the risk factors which characterize longer power outages? 

To answer these questions, we analyze `power`, which is described below.

## Step 2: Data Cleaning and Exploratory Data Analysis

```python
import pandas as pd
import numpy as np
from pathlib import Path

import plotly.express as px
pd.options.plotting.backend = 'plotly'
import folium

from dsc80_utils import * # Feel free to uncomment and use this.
```

```python
#Read in the data
path = "outage_data.xlsx"
power = pd.read_excel(path, header=None)

#Drop dummy header and keep the true header
power = power[5:].copy()          
power.columns = power.iloc[0]   
power = power[1:]                
power = power.reset_index(drop=True)

#Drop unnecessary columns
power.drop(columns=["variables"], inplace=True)
power = power.iloc[1:].reset_index(drop=True)
```

```python
#datetime cleaning
power['OUTAGE.START'] = (
    pd.to_datetime(power['OUTAGE.START.DATE']) +
    pd.to_timedelta(power['OUTAGE.START.TIME'].astype(str), errors='coerce')
)

power['OUTAGE.RESTORATION'] = (
    pd.to_datetime(power['OUTAGE.RESTORATION.DATE']) +
    pd.to_timedelta(power['OUTAGE.RESTORATION.TIME'].astype(str), errors='coerce')
)
# power.columns
```

As an initial visual demonstration, we're intrerested in seeing the visual spread of these outages across the U.S.

```python
counts = power.groupby('POSTAL.CODE').size().reset_index(name='outages')
counts
```

```python
fig = px.choropleth(
    counts,
    locations='POSTAL.CODE',       
    locationmode="USA-states",     
    color='outages',              
    scope="usa",                   
    color_continuous_scale="Reds",
    labels={'outages': 'Power Outages'},
)

fig.update_traces(marker_line_color='white')  # add white borders between states
# Add a title
fig.update_layout(
    title_text='Power Outages by State in the U.S.',
    title_x=0.5,
    title_font_size = 24,
    margin=dict(t=100, l=70, r=40, b=70)
    
)

fig.show()
```

```python
cause_props = power['CAUSE.CATEGORY'].value_counts(normalize = True).reset_index()

cause_props = power['CAUSE.CATEGORY'].value_counts(normalize=True).reset_index()
cause_props.columns = ['CAUSE.CATEGORY', 'proportion']

# Create bar chart
fig = px.bar(
    cause_props, 
    x='CAUSE.CATEGORY', 
    y='proportion',
    labels={
        'CAUSE.CATEGORY': 'Cause Category',
        'proportion': 'Proportion'
    },
    title="Distribution of Cause Category Proportions",
    color_discrete_sequence=['salmon']  # single color for all bars
)

# Update layout with larger margins to prevent title cutoff
fig.update_layout(
    title_font=dict(size=24),
    xaxis=dict(title=dict(font=dict(size=18))),
    yaxis=dict(title=dict(font=dict(size=18))),
    autosize=False,
    width=800,
    height=600,
    margin=dict(t=100, l=70, r=40, b=70)  # top margin increased to 100
)

fig.show()
```

```python
fig = px.histogram(
    power,
    x='OUTAGE.DURATION',
    nbins=50,
    title='Distribution of Outage Duration',
    labels={'OUTAGE.DURATION': 'Outage Duration'},
    color_discrete_sequence=['salmon']
)

fig.update_layout(
    
    xaxis_title='Outage Duration',
    yaxis_title='Count',
    bargap=0.1,
    margin=dict(t=60, l=50, r=30, b=50)
)
fig.show()
```
